```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as p
```

```python
df=pd.read_csv('C:\\Users\\gondu\\Downloads\\archive (5)\\Titanic-Dataset.csv')
df.info()
```

```python
df.drop(columns=['Name','Cabin'],inplace=True)
```

```python
df['Age'].fillna(df['Age'].mean,inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mean,inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Sex']=encoder.fit_transform(df['Sex'])
```

```python
df.isna().sum()
```

```python
df["Survived"].value_counts().plot(kind="bar", color=["red", "green"])
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Survived", data=df, palette=["red", "green"])
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Pclass", data=df, palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x="Embarked", hue="Survived", data=df, palette=["red", "green"])
plt.xlabel("Embarked Location")
plt.ylabel("Count")
plt.title("Survival Rate by Embarkation Point")
plt.show()
```

```python
plt.hist(df["Fare"], bins=30, color="purple", edgecolor="black")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Fare Distribution")
plt.show()
```
